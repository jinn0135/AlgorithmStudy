```python
# 스택: 괄호
# https://www.acmicpc.net/problem/9012
def VPS(item):
    stack = []
    for i in item:
        if i=='(': stack.append(i)
        elif i==')': 
            if len(stack)==0: return 'NO'
            stack.pop()
    if len(stack)==0: return 'YES'
    else: return 'NO'

num = int(input())
for i in range(num):
    item = input()
    print(VPS(item))
    
# 6
# (())()) # NO
# (((()())() # NO
# (()())((())) # YES
# ((()()(()))(((())))() # NO
# ()()()()(()()())() # YES
# (()((())()( # NO

# 3
# (( # NO
# )) # NO
# ())(() #NO
```

```python
# 스택: 크레인 인형뽑기
# https://school.programmers.co.kr/learn/courses/30/lessons/64061
def solution(board, moves):
    ans = 0; stack = []
    for m in moves:
        if board[-1][m-1]==0: continue
        for h in range(len(board)):
            if board[h][m-1]!=0:
                stack.append(board[h][m-1])
                board[h][m-1] = 0
                while len(stack)>1 and stack[-1]==stack[-2]:
                    stack.pop()
                    stack.pop()
                    ans += 2
                break
    return ans

board = [[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]]
moves = [1,5,3,5,1,2,1,4] # ans=4
```

```python
# 큐: 카드2
# https://www.acmicpc.net/problem/2164
from collections import deque
def card(num):
    que = deque([i for i in range(1,num+1)])
    while len(que)>2:
        que.popleft()
        que.append(que.popleft())
    
    if len(que)==2:
        que.popleft()
    return que.pop()

num = int(input())
ans = card(num)
print(ans)

num = 6 # ans=4
```

```python
# 큐: 프린터
# https://school.programmers.co.kr/learn/courses/30/lessons/42587
from collections import deque
def solution(priorities, location):
    pri_que = deque(priorities)
    loca_que = deque([0 for i in range(len(priorities))])
    loca_que[location] = 1
    
    while len(loca_que)!=0 and max(loca_que)==1:
        if pri_que[0]==max(pri_que):
            pri_que.popleft()
            loca_que.popleft()
        else:
            pri_que.append(pri_que.popleft())
            loca_que.append(loca_que.popleft())
    return len(priorities)-len(loca_que)

priorities = [2, 1, 3, 2]; location = 2 # ans=1
priorities = [1, 1, 9, 1, 1, 1]; location = 0 # ans=5
```

```python
# 재귀함수: 괄호변환
# https://school.programmers.co.kr/learn/courses/30/lessons/60058
from collections import deque
def solution(p):
    if p=='': return p

    a_n, b_n = 0, 0
    for i in range(len(p)):
        if p[i]=='(': a_n += 1
        elif p[i]==')': b_n += 1
        if a_n!=0 and a_n==b_n:
            u, v = p[:i+1], p[i+1:]
            break
    if right_p(u): return u+solution(v)
    
    ans = '('+solution(v)+')'
    for s in u[1:-1]:
        if s=='(': ans += ')'
        elif s==')': ans += '('
    return ans

def right_p(p):
    if p=='': return True
    p, stack = deque(p), deque([])
    while p:
        s = p.popleft()
        if s=='(': stack.append(s)
        elif s==')':
            if len(stack)==0:
                return False
            stack.pop()
    if len(p)==0: return True
    else: return False
    
p, ans = "(()())()", "(()())()"
p, ans = ")(", "()"
p, ans = "()))((()", "()(())()"
```

```python
# 트리/그래프: 트리의 부모찾기
# https://www.acmicpc.net/submit/11725/59508300
from collections import deque
def func(count, nums):
    visit = deque([1])
    ans = {i:0 for i in range(1,count+1)}
    ans[1] = 1
    while visit:
        now = visit.popleft()
        for child in nums[now]:
            if child!=1 and ans[child]==0:
                ans[child] = now     
                visit.append(child)
    for i in range(2,count+1):
        print(ans[i])

count = int(input())
nums = {i:[] for i in range(1,count+1)}
for i in range(count-1):
    a,b = map(int, input().split(' '))
    nums[a].append(b)
    nums[b].append(a)  
func(count, nums)
```

```python
# 트리/그래프: 순위
# https://school.programmers.co.kr/learn/courses/30/lessons/49191
import numpy as np
def solution(n, results):
    if n==1 or n==2: return n
    graph = [[np.nan for i in range(n)] for i in range(n)]
    for i in range(len(graph)):
        graph[i][i] = 0
    for a,b in results:
        graph[a-1][b-1] = 1 # win
        graph[b-1][a-1] = -1 # lose
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                if graph[i][k]==1 and graph[k][j]==1:
                    graph[i][j] = 1
                    graph[j][i] = -1
    for i in range(n):
        for j in range(n):
            for k in range(n):
                if graph[i][k]==1 and graph[k][j]==1:
                    graph[i][j] = 1
                    graph[j][i] = -1                
    
    ans = 0
    for i in range(n):
        if np.isnan(graph[i]).sum()==0:
            ans += 1
#     print(graph)
    return ans

n, results, ans = 5, [[4, 3], [4, 2], [3, 2], [1, 2], [2, 5]], 2
n, results, ans = 7, [[1,2],[1,4],[1,5],[2,3],[5,6],[3,7],[4,7],[6,7]], 2
n, results, ans = 6, [[1,2],[1,4],[2,3],[4,5],[3,6],[5,6]], 2
n, results, ans = 7, [[1,2],[2,3],[3,4],[4,7],[2,5],[5,6],[6,7]], 3
n, results, ans = 5, [[1,3], [3,5], [5,4], [4,2]], 5
solution(n, results)
```

```python
# 트리/그래프: 가장 먼 노드
# https://school.programmers.co.kr/learn/courses/30/lessons/49189
from collections import deque
def solution(n, vertex):
    layer = {i:0 for i in range(1,n+1)}
    graph = {i:[] for i in range(1,n+1)}
    for a,b in vertex:
        graph[a].append(b)
        graph[b].append(a)
    visit = deque([1])
    while visit:
        now = visit.popleft()
        for node in graph[now]:
            visit.append(node)
            if layer[node]==0:
                layer[node] = layer[now]+1
            else: 
                layer[node] = min(layer[now]+1, layer[node])
            graph[node].remove(now)
        graph[now] = []
    max_layer = max(layer.values())
    ans = 0
    for k,v in layer.items():
        if v==max_layer:
            ans+=1
    return ans

n, vertex, ans = 6, [[3,6],[4,3],[3,2],[1,3],[1,2],[2,4],[5,2]], 3
solution(n, vertex)
```

```python
# 완전탐색/이분탐색: 연산자 끼워넣기
# https://www.acmicpc.net/problem/14888
count = int(input())
nums = list(map(int, input().split(' ')))
get = list(map(int, input().split(' '))) # +, -, *, /
max_v, min_v = -1e9, 1e9
def func(i, num, count=count, nums=nums):
    global max_v, min_v, get
    if i==count:
        max_v = max(max_v, num)
        min_v = min(min_v, num)
    else:
        if get[0]: # '+'
            get[0] -= 1
            func(i+1, num+nums[i])
            get[0] += 1
        if get[1]: # '-'
            get[1] -= 1
            func(i+1, num-nums[i])
            get[1] += 1
        if get[2]: # '*'
            get[2] -= 1
            func(i+1, num*nums[i])
            get[2] += 1
        if get[3]: # '/'
            get[3] -= 1
            func(i+1, int(num/nums[i]))
            get[3] += 1

func(1, nums[0])
print(max_v)
print(min_v)
```

```python
# 완전탐색/이분탐색: 체스판 다시 칠하기
# https://www.acmicpc.net/problem/1018
def func(n, m, board):
    ans = 50*50
    for i in range(n-7):
        for j in range(m-7):
            ans = min(ans, WBboard(i,j,board), BWboard(i,j,board))
    return ans
def WBboard(i,j,board):
    s1 = 'WB'*4; s2 = 'BW'*4
    check = [s1, s2, s1, s2, s1, s2, s1, s2]
    miss = 0
    for check_i, board_i in zip(range(8), range(i,i+9)):
        if check[check_i]==board[board_i]: continue
        for check_j, board_j in zip(range(8), range(j,j+9)):
            if check[check_i][check_j]!=board[board_i][board_j]:
                miss += 1
    return miss
def BWboard(i,j,board):
    s1 = 'BW'*4; s2 = 'WB'*4
    check = [s1, s2, s1, s2, s1, s2, s1, s2]
    miss = 0
    for check_i, board_i in zip(range(8), range(i,i+9)):
        if check[check_i]==board[board_i]: continue
        for check_j, board_j in zip(range(8), range(j,j+9)):
            if check[check_i][check_j]!=board[board_i][board_j]:
                miss += 1
    return miss

# n, m = map(int, input().split(' '))
# board = []
# for i in range(n):
#     board.append(input())
n,m,board = 8, 8, ['WBWBWBWB','BWBWBWBW','WBWBWBWB','BWBBBWBW',
                   'WBWBWBWB','BWBWBWBW','WBWBWBWB','BWBWBWBW'] # 1
n,m,board = 10, 13, ['BBBBBBBBWBWBW','BBBBBBBBBWBWB','BBBBBBBBWBWBW',
                     'BBBBBBBBBWBWB','BBBBBBBBWBWBW','BBBBBBBBBWBWB',
                     'BBBBBBBBWBWBW','BBBBBBBBBWBWB','WWWWWWWWWWBWB',
                     'WWWWWWWWWWBWB'] # 12
ans = func(n,m, board)
print(ans)
```

```python
# BFS/DFS: DFS와 BFS
# https://www.acmicpc.net/problem/1260    
def dfs(v,nodes):
    visit1 = {v:1}
    stack = [v]
    while stack:
        now = stack.pop()
        visit1[now] = 1
        if now not in nodes:
            continue
        else:
            nodes[now] = sorted(nodes[now], reverse=True)
            for node in nodes[now]:
                if node not in visit1:
                    stack.append(node)
    return visit1

def bfs(v,nodes):
    visit2 = {v:1}
    if v not in nodes: return visit2
    que = [sorted(nodes[v])]
    while que:
        for i in range(len(que)):
            now = que.pop(0)
            for node in nodes[now]:
                if node not in visit2:
                    que.append(node)
    
    nodes[v] = sorted(nodes[v])
    for node in nodes[v]:
        if node not in visit2:
            stack.append(node)
            visit2[node] = 1
    if len(stack)==0: return None
    for s in stack:
        bfs(s,nodes)

n, m, v = 4, 5, 1
nodes = {1:[2,3,4],2:[1,4],3:[1,4],4:[1,2,3]} # 1243 / 1234
n, m, v = 5, 5, 3
nodes = {1:[2,3],2:[5,1],3:[4,1],4:[5,3],5:[4,2]} # 31254 / 31425
n, m, v = 1000, 1, 1000
nodes = {999:[1000],1000:[999]} # 1000 999 / 1000 999
# n, m, v = map(int, input().split(' '))
# nodes = {i:[] for i in range(1,n+1)}
# for i in range(m):
#     a,b = map(int, input().split(' '))
#     nodes[a].append(b)
#     nodes[b].append(a)
visit1 = dfs(v,nodes)
print(*visit1)
bfs(v,nodes)
print(*visit2)
```
